```python
def trd(element):
    if pd.isnull(element): pass
    elif element<3:
        return 'New'
    elif element >= 3 and element < 6:
        return 'Experienced'
    elif element >= 7 and element < 10:
        return 'Established'
    elif element > 11:
        return 'Veteran'
        
    
```

```python
combined_updated['service_cat'] = combined_updated['institute_service'].apply(trd)
```
